```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
# Set the random seed for reproducibility
np.random.seed(3)

# Generate 1000 random numbers from a normal distribution
mean = 150  # Mean of the distribution
std_dev = 50  # Standard deviation of the distribution
normal_values = np.random.normal(mean, std_dev, 1000)

# Scale and shift the values to fit within the range [1, 300]
scaled_values = np.clip(normal_values, 1, 300)

# Round the values to the nearest integer
integer_values = np.round(scaled_values).astype(int)

# Print the first 10 values to verify
print(integer_values)
```

```python
# Select a random sample of 100 integers from the generated 1000 integers
def sampling():
    sample_values = np.random.choice(integer_values, 100, replace=False)
    return sample_values.mean(), sample_values.var(), (sample_values.mean()+(2/100))
```

```python

def plot_histograms(data):
    # Sample data (replace these with your actual lists)
    data1 = data
    # data2 = mu_hat_vec

    # Create histograms
    counts1, bins1 = np.histogram(data1, bins=30)
    # counts2, bins2 = np.histogram(data2, bins=30)

    # Calculate the bin centers
    bin_centers1 = 0.5 * (bins1[1:] + bins1[:-1])
    # bin_centers2 = 0.5 * (bins2[1:] + bins2[:-1])

    # Plot histograms
    # plt.hist(data2, bins=30, alpha=0.5, label='Mu Hat', color='red')
    plt.hist(data1, bins=30, alpha=0.5, label='Mean', color='blue')

    # Plot lines connecting the top points of the histograms
    # plt.plot(bin_centers2, counts2, '-o', color='red')
    plt.plot(bin_centers1, counts1, '-o', color='blue')

    # Add labels and title
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Frequency Distribution ')
    plt.legend(loc='upper right')

    # Show plot
    plt.show()
```

```python
mean_vec,var_vec,mu_hat_vec = np.zeros((200)),np.zeros((200)),np.zeros((200))
for i in range(200):
    x=sampling()
    mean_vec[i]=x[0]
    var_vec[i]=x[1]
    mu_hat_vec[i]=x[2]
plot_histograms(mean_vec)
plot_histograms(mu_hat_vec)
plot_histograms(var_vec)
```

```python
mean_vec_2,var_vec_2,mu_hat_vec_2 = np.zeros((500)),np.zeros((500)),np.zeros((500))
for i in range(500):
    x=sampling()
    mean_vec_2[i]=x[0]
    var_vec_2[i]=x[1]
    mu_hat_vec_2[i]=x[2]
plot_histograms(mean_vec_2)
plot_histograms(mu_hat_vec_2)
plot_histograms(var_vec_2)
```

```python
def plot_histograms_multi(data_a,data_b):
    data1 = data_a
    data2 = data_b

    counts1, bins1 = np.histogram(data1, bins=30)
    counts2, bins2 = np.histogram(data2, bins=30)

    bin_centers1 = 0.5 * (bins1[1:] + bins1[:-1])
    bin_centers2 = 0.5 * (bins2[1:] + bins2[:-1])

    plt.hist(data2, bins=30, alpha=0.5, label="data2", color='red')
    plt.hist(data1, bins=30, alpha=0.5, label="data1", color='blue')

    plt.plot(bin_centers2, counts2, '-o', color='red')
    plt.plot(bin_centers1, counts1, '-o', color='blue')

    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Frequency Distribution of Two Datasets')
    plt.legend(loc='upper right')

    plt.show()
```
