```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact, IntSlider

# Set up the normal distribution parameters
mean_u = 0
std_u = 4  # Since the variance is 16, the standard deviation is sqrt(16) = 4

def sample_statistics(n, num_repeats=200):
    sample_means = []
    sample_variances = []
    mus = []
    
    for _ in range(num_repeats):
        # Draw a sample of size n
        random_floats = np.random.normal(mean_u, std_u, n)

        # Round the numbers to the nearest integer
        u_samples = np.round(random_floats).astype(int)
        X_samples = 5 + u_samples
        
        # Calculate sample mean and variance
        sample_mean = np.mean(X_samples)
        sample_variance = np.var(X_samples, ddof=1)  # ddof=1 for sample variance
        
        # Calculate mu
        mu = sample_mean + (2 / n)
        
        # Store the statistics
        sample_means.append(sample_mean)
        sample_variances.append(sample_variance)
        mus.append(mu)
    
    return sample_means, sample_variances, mus
```

```python
sample_statistics(100)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample list of floating point numbers
data = sample_statistics(100)[0]  # Replace this with your list of numbers

# Create histogram with 8 bins
counts, bin_edges = np.histogram(data, bins=8)

# Calculate the midpoints of the bins
bin_midpoints = (bin_edges[:-1] + bin_edges[1:]) / 2

# Plot the histogram with bar plot
plt.hist(data, bins=8, edgecolor='black', alpha=0.5)

# Plot the line connecting the midpoints of the tops of the bars
plt.plot(bin_midpoints, counts, marker='o', linestyle='-', color='red')

# Add labels and title for clarity
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Frequency Plot with Lines Connecting Bar Midpoints')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample list of floating-point numbers
data = np.random.randn(100)  # Replace this with your list of numbers

# Create histogram with 8 bins
counts, bin_edges = np.histogram(data, bins=8)

# Calculate the midpoints of the bins
bin_midpoints = (bin_edges[:-1] + bin_edges[1:]) / 2

# Create a new figure and axis
fig, ax = plt.subplots()

# Plot the histogram with bar plot
ax.hist(data, bins=8, edgecolor='black', alpha=0.5, label='Histogram')

# Plot the line connecting the midpoints of the tops of the bars
ax.plot(bin_midpoints, counts, marker='o', linestyle='-', color='red', label='Line Plot')

# Create another plot (for demonstration, we'll use a sine wave)
x = np.linspace(min(data), max(data), 100)
y = np.sin(x)

# Plot the second plot on the same graph
ax.plot(x, y, color='blue', linestyle='--', label='Sine Wave')

# Add labels, title, and legend for clarity
ax.set_xlabel('Value')
ax.set_ylabel('Frequency / Value')
ax.set_title('Frequency Plot with Additional Plot')
ax.legend()

# Show the plot
plt.show()
```
